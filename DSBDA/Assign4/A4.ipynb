```python
outliers = df.quantile(.97) # Dealing with the outliers
df = df[(df['CRIM']<outliers['CRIM'])]
df = df[(df['ZN']<outliers['ZN'])]
df = df[(df['RM']<outliers['RM'])]
df = df[(df['DIS']<outliers['DIS'])]
df = df[(df['PTRATIO']<outliers['PTRATIO'])]
df = df[(df['B']<outliers['B'])]
df = df[(df['LSTAT']<outliers['LSTAT'])]
```

```python
abs(df.corr()["MEDV"].sort_values(ascending=False))
```

```python
x,y=df.drop("MEDV",axis=1),df[["MEDV"]]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error

L=LinearRegression() 

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)

result=pd.DataFrame(columns=['R_Squared','RMSE','MAE'])

r_squared=[]
rmse=[]
mae=[]

L.fit(x_train, y_train)
L.predict(x_test)

print(L.predict(x_test))

r_squared.append(r2_score(y_test,L.predict(x_test)))
rmse.append((mean_squared_error(y_test,L.predict(x_test)))**.5)
mae.append(mean_absolute_error(y_test,L.predict(x_test)))
        
result.R_Squared=r_squared
result.RMSE=rmse
result.MAE=mae


print(result.sort_values('R_Squared',ascending=False))
```
